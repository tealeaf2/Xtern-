<h1>Exploratory Analysis of Data<h1>

```python
#Putting into dataframe
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
df = pd.read_csv("Xtern_traindata.csv")
print(df)
```

<h3>Looking at trends of Orders in comparison to all other factors<h3>

```python
df.plot(kind="scatter", x = "Year", y = "Order")
plt.xticks(rotation=45, ha='right')
plt.show()
```

Analysis: Year 2 and Year 3 students generally do not have a preference on orders while Year 1 and Year 4 order very few things. Specifically, Year 4 students only orders tacos.

```python
df.plot(kind="scatter", x = "Major", y = "Order")
plt.xticks(rotation=45, ha='right')
plt.show()
```

Analysis: Looking at majors in comparison to orders, we can see that from Finance and onwards, orders tend to be more specified for the specific major. As an example, Physics and Chemistry majors tend to have no preference in orders while Fine Arts and Civil Engineering majors only order 2 items from the list.

```python
df.plot(kind="scatter", x = "University", y = "Order")
plt.xticks(rotation=45, ha='right')
plt.show()
```

Analysis: For universities, we can see that Indiana State, Ball State, Butler University, and Notre Dame orders all of the items on the menu. In comparison to Valparaiso, Purdue, Indiana University Bloomington, and DePauw, there is a lot fewer orders as well as each university having a prefered choice of orders in that location.

```python
df.plot(kind="scatter", x = "Time", y = "Order")
plt.xticks(rotation=45, ha='right')
plt.show()
```

Analysis: Time is an important factor for orders because some orders are only seen in certain timeframes. For example, Hoosier BBQ Sandwich is only seen ordered during the time of 11-15 while Cornbread Hush Puppies are ordered starting at 8 and stopping at 15.

<h3>Distribution<h3>

```python
letter_counts = Counter(df["Major"])
df_major = pd.DataFrame.from_dict(letter_counts, orient='index')
df_major.plot(kind='bar')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Distribution of majors is not in any shape or form, but bias can be seen to veer towards majors with high frequency, such as Physics and Chemistry.

```python
letter_counts = Counter(df["Year"])
df_year = pd.DataFrame.from_dict(letter_counts, orient='index')
df_year.plot(kind='bar')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Distribution is heavily skewed towards Year 2 and Year 3, meaning that bias will be towards these years.

```python
letter_counts = Counter(df["University"])
df_uni = pd.DataFrame.from_dict(letter_counts, orient='index')
df_uni.plot(kind='bar')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Indiana State, Ball State, and Butler will have a significantly bias in the prediction of order based on the frequency of orders in these colleges. This skews the distribution.

```python
letter_counts = Counter(df["Order"])
df_order = pd.DataFrame.from_dict(letter_counts, orient='index')
df_order.plot(kind='bar')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Distribution of orders is even, meaning that orders will not matter much as a factor and can be treated as unbiased.

```python
df["Time"].plot(kind="hist")
plt.show()
```

Distribution of orders is in a bell curve, where the peak is between the times of 11-14. This means that bias is significantly veered towards this time period.

Business:

For higher sales, the data suggests that more efforts should be focused:

<h4>Correlation between all variables<h4>

```python
df['Year']=df['Year'].astype('category').cat.codes
df['Major']=df['Major'].astype('category').cat.codes
df['University']=df['University'].astype('category').cat.codes
df['Order']=df['Order'].astype('category').cat.codes
print(df.corr())
```

Analysis: Linear relationships can be slightly found between Order/Time, University/Time, and Order/Year. These factors seem to be where biases will be weighted most highly, as they matter more in predicting the order than any other variables.
